# 18  ta CSV fayllarni bitta yagona datasetga birlashtirib, uni yangi fayl sifatida saqlaymiz.

```python
import pandas as pd
import os

# üìÅ Fayllar joylashgan joy: Merging_Data
source_folder = r"C:\Users\Rasulbek907\Desktop\Final_Project\Data\Merging_Data"

# üìÑ Fayllar ro'yxati
filenames = [f"hf_dataset{i}.csv" for i in range(1, 19)]

# üì• CSV fayllarni o'qib birlashtirish
dataframes = []
for file in filenames:
    file_path = os.path.join(source_folder, file)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        dataframes.append(df)
        print(f"‚úÖ Yuklandi: {file}")
    else:
        print(f"‚ùå Topilmadi: {file}")

# üß© Barchasini birlashtirish
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)

    # üì§ Yangi faylni saqlash manzili (Raw_data papkasi)
    output_path = r"C:\Users\Rasulbek907\Desktop\Final_Project\Data\Raw_data\merged_dataset.csv"
    merged_df.to_csv(output_path, index=False, encoding='utf-8')

    print(f"\n‚úÖ Yangi fayl saqlandi: {output_path}")
else:
    print("\n‚ö†Ô∏è Hech qanday fayl topilmadi yoki yuklanmadi.")
```
